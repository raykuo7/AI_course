## 6.3.3 One-Hot 編碼法

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('客戶聯絡狀況資料檔.csv')
label1 = LabelEncoder()
df['訂購'] = label1.fit_transform(df['訂購'])
onehot = OneHotEncoder(sparse_output=False)
arr = onehot.fit_transform(df[['訂購']])
arr
```

# 6.6 使用Scikit-Learn 進行特徵選擇

##載入Scikit-Learn 的卡方驗證模組

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

##取出判斷特徵與目標特徵

1.建立判斷特徵

```python
df = pd.read_csv('/content/drive/MyDrive/資料科學自學聖經/ch06/乳癌病歷資料檔.csv')
df
```

```python
data = df.iloc[:, :-1]
data
```

2.建立目標特徵

```python
target = df.iloc[:, -1]
target
```

##進行卡方驗證

```python
n = 9
chi = SelectKBest(chi2, k=n)
arrchi = chi.fit_transform(data, target)
arrchi
```

##取得特徵驗證值

```python
score = chi.scores_
score
```

##取得相關特徵名稱

```python
scoresort = np.argsort(score)
scoresort = np.flipud(scoresort)
scoresort
```

使用迴圈由索引值取得特徵名稱並顯示

```python
col = df.columns
print('選擇的特徵：')
for i in range(n):
  print('{}：{}'.format(col[scoresort[i]], score[scoresort[i]]))
```
