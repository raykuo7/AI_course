## 5.1.4 異常值處理

```python
# df = pd.read_csv('/content/drive/MyDrive/資料科學自學聖經/ch05/學生月考成績檔.csv')
df = pd.read_csv('學生月考成績檔.csv')
df = df.dropna()
colname = '國文'
Q1 = df[colname].quantile(0.25)
Q3 = df[colname].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
df1 = df[(df[colname]<lower) | (df[colname]>upper)]
print("Q1:",Q1,"Q3:",Q3,"IQR:",IQR,"upper:",upper,"lower:",lower)
print(df1.shape[0])
df1
```

## 5.5.1 keras ImageDataGenerator 模組

```python
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import cv2

imgGen = ImageDataGenerator(shear_range=50.0, fill_mode='reflect')
n = 5
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR轉成RGB
img_sr = img.copy()
img = np.array(img, dtype=np.float32)
img = np.expand_dims(img, 0) #輸入圖片要是四維
flowGen = imgGen.flow(img, batch_size=10)

plt.figure(figsize=(20,10))
plt.subplot(1, n+1, 1)
plt.imshow(img_sr)
for i in range(n):
  img = flowGen[0][0, :, :, :].astype(np.uint8)
  plt.subplot(1, n+1, i+2)
  plt.imshow(img)
  plt.axis('off')
```

```python
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import cv2

imgGen = ImageDataGenerator(shear_range=50.0, fill_mode='reflect')
n = 5
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR轉成RGB
img = np.array(img, dtype=np.float32)
img = np.expand_dims(img, 0) #輸入圖片要是四維
flowGen = imgGen.flow(img, save_to_dir='.',
                      save_prefix='cat',
                      save_format='jpg',
                      batch_size=10)

for i in range(n):
  img = flowGen[0][0, :, :, :].astype(np.uint8)
```

```python
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import cv2

imgGen = ImageDataGenerator(
    rotation_range=60.0,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=50,
    zoom_range=0.5,
    horizontal_flip=True,
    fill_mode='reflect',)
n = 5
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR轉成RGB
img_sr = img.copy()
img = np.array(img, dtype=np.float32)
img = np.expand_dims(img, 0) #輸入圖片要是四維
flowGen = imgGen.flow(img, batch_size=10)

plt.figure(figsize=(20,10))
plt.subplot(1, n+1, 1)
plt.imshow(img_sr)
for i in range(n):
  img = flowGen[0][0, :, :, :].astype(np.uint8)
  plt.subplot(1, n+1, i+2)
  plt.imshow(img)
  plt.axis('off')
  img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) #RGB轉成BGR
  cv2.imwrite('cat_' + str(i+1) + '.jpg', img)
```
